Säuglingssterblichkeit für zwei Zweitintervalle mit Länderaufschlüsselung.
Quelle: http://www.gbe-bund.de/gbe10/ergebnisse.prc_tab?fid=13853&suchstring=sterblichkeit&query_id=&sprache=D&fund_typ=TAB&methode=2&vt=1&verwandte=1&page_ret=0&seite=&p_lfd_nr=28&p_news=&p_sprachkz=D&p_uid=gast&p_aid=23367600&hlp_nr=3&p_janein=J#SEARCH=%2522sterblichkeit%2522
1. beiden Spalten: Säuglingssterbefälle pro 1.000 Lebendgeborene 2006 bis 2008
2. beiden Spalten: Veränderung von 1991 bis 1993 bis 2006 bis 2008 

```python
# imports
import pandas as pd
```

```python
columns = ['Land', 'weiblich_2008', 'männlich_2008', 'weiblich_diff', 'männlich_diff']
data = [
    ['Schleswig-Holstein', 3.2, 3.8, -37, -45],
    ['Hamburg', 3.1, 3.5, -46, -56],
    ['Niedersachsen', 3.5, 5.2, -39, -19],
    ['Bremen', 4.1, 6.3, -11, 2],
    ['Nordrhein-Westfalen', 4.2, 4.9 ,-30, -36],
    ['Hessen', 3.8, 4.0, -23, -43],
    ['Rheinland-Pfalz', 3.7, 4.3, -36, -41 ],
    ['Baden-Württemberg', 2.7, 3.5, -45, -41 ],
    ['Bayern', 2.9, 3.9, -40, -38],
    ['Saarland', 3.7, 4.4, -36, -47 ],
    ['Berlin', 3.2, 3.8, -47, -45],
    ['Brandenburg', 2.5, 3.0,  	-56, -65],
    ['Mecklenburg-Vorpommern', 3.1, 3.7, -59, -53 ],
    ['Sachsen', 2.5, 2.6, -57, -65 ],
    ['Sachsen-Anhalt', 3.3, 3.3, -51, -61 ],
    ['Thüringen', 3.1, 4.1, -56, -54 ]
]
df_sterblichkeit = pd.DataFrame(data=data, columns=columns)
```

```python
# new columns
df_sterblichkeit['weiblich_diff'] = df_sterblichkeit['weiblich_diff']/100-1
df_sterblichkeit['männlich_diff'] = df_sterblichkeit['männlich_diff']/100-1
df_sterblichkeit['weiblich_1993'] = df_sterblichkeit['weiblich_2008'] * (-1) * df_sterblichkeit['weiblich_diff']
df_sterblichkeit['männlich_1993'] = df_sterblichkeit['männlich_2008'] * (-1) * df_sterblichkeit['männlich_diff']
```

```python
batches = ['weiblich_2008', 'männlich_2008', 'weiblich_1993', 'männlich_1993']
df_2008 = df_sterblichkeit[['Land', 'weiblich_2008', 'männlich_2008']]
df_2008['year'] = 2008
df_2008 = df_2008.rename(columns={'weiblich_2008': 'weiblich', 'männlich_2008': 'männlich'}, errors='raise')
df_1993 = df_sterblichkeit[['Land', 'weiblich_1993', 'männlich_1993']]
df_1993['year'] = 1993
df_1993 = df_1993.rename(columns={'weiblich_1993': 'weiblich', 'männlich_1993': 'männlich'})
df_new = df_2008.append(df_1993)
```
